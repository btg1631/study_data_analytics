## conditional filtering

```python
import pandas as pd
# DataFrame 생성
data = {
    '이름': ['홍길동', '김철수', '이영희', '박지성', '김연아'],
    '나이': [25, 32, 19, 28, 31],
    '성별': ['남', '남', '여', '남', '여']
}
df_temp = pd.DataFrame(data)
df_temp
```

### boolean indexing

```python
# '나이 >= 30'
# df_temp[df_temp['나이'] >= 30]
df_temp.loc[df_temp['나이'] >= 30]
```

```python
condition_males = df_temp['성별'] == '남'
condition_under_30 = df_temp['나이'] < 30
conditions = condition_males & condition_under_30
df_temp.loc[conditions]
```

```python
df_temp.set_index('이름', inplace=True) # '이름' 컬럼을 index로 지정
df_temp
```

```python
condition_in = ['홍길동', '이영희']
df_temp.loc[["홍길동", "이영희"]]
df_temp.loc[condition_in]
```

### query

```python
df_temp.query('나이 >= 30')
```

```python
condition_males = '성별 == "남"'
df_temp.query(condition_males)
```

```python
df_temp[['이름', '성별']].query(condition_males)
# df_temp[['이름', '성별']].query(condition_males).columns
```

```python
condition_under_30 = '나이 < 30'
# conditions = '{} and {}'.format(condition_males, condition_under_30)
conditions = f'{condition_males} and {condition_under_30}'
# conditions
df_temp.query(conditions)
```

```python
condition_in = '이름 in ["홍길동", "이영희"]'
df_temp.query(condition_in)
```
