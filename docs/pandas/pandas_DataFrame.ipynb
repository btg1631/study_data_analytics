#### DataFrame 생성

```python
list = [3, 4, 5, 6]
pd.DataFrame(list)
```

```python
double_list = [
             [1000, '과자','2019-12-31','반품'],
             [2000, '음료', '2020-03-02', '정상'],
             [3000, '아이스크림', '2020-02-03','정상'],
             [1000,'과자','2019-12-31','반품']
            ]
pd.DataFrame(data = double_list)
```

```python
double_columns = ['가격','종류','판매일자','반품여부']
pd.DataFrame(data = double_list, columns = double_columns)
```

#### row(index) 단위 : iloc vs loc
- iloc : 위치값으로 접근
- loc : 명칭 접근

```python
df_saledays
```

```python
df_saledays.index = ['a', 'b', 'c', 'd']
```

```python
df_saledays.iloc[2]
df_saledays.loc['c']
```

#### index와 columns 같이 추출

```python
df_saledays.loc['c', '판매일자']  # indexes, columns
```

```python
df_saledays.loc[:, '판매일자']  # indexes, columns
```

```python
df_saledays.columns
```

```python
df_saledays.loc[:, '종류':]
```

```python
df_saledays.loc[:, '판매일자':]  # indexes, columns
```
